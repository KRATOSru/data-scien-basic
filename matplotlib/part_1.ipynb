#### Matplotlib 2

```python
figure, axes = plt.subplots(nrows=2, ncols=4)
plt.tight_layout()
# axes.plot(a, b)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
        current_ax.plot(a, b)
```

```python
axes
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(a, b)

axes[0].set_xlabel("a array")
axes[0].set_ylabel("b array")
axes[0].set_title("Axeas[0] Title")

axes[1].plot(b, a)

axes[1].set_xlabel("b array")
axes[1].set_ylabel("a array")
axes[1].set_title("Axeas[1] Title")
```

### Figure size , aspect ratio and dpi

```python
figure = plt.figure(figsize=(7, 2))
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a, b)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=1, figsize=(7, 2))
axes[0].plot(a, b)
axes[1].plot(b, a)
plt.tight_layout()
```

```python
figure
```

```python
figure.savefig('m_plot.png')
```

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a, b, "r", label="a, b plot")
axes.plot(b, a, label="b, a plot")
axes.legend(loc=0)
```

### Matplotlib 3

```python
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(
    a,
    b,
    color="#e5960d",
    linewidth=5,
    alpha=0.3,
    linestyle="-.",
    marker="o",
    markersize=10,
    markerfacecolor="green",
    markeredgewidth=3,
    markeredgecolor="red",
)
```
