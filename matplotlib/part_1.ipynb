OOP Method

```python
figure = plt.figure()
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(a, b, 'g')
axes.set_xlabel('a array')
axes.set_ylabel('b array')
axes.set_title('A Simple Plot')
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = figure1.add_axes([0.3, 0.6, 0.5, 0.4])
```

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0.0, 0.0, 1, 1])
axes2 = figure1.add_axes([0.1, 0.4, 0.5, 0.4])

axes1.plot(a,b)
axes2.plot(b,a,'r')

axes1.set_xlabel('a array')
axes1.set_ylabel('b array')
axes1.set_title('Axeas1 Title')

axes2.set_xlabel('a array')
axes2.set_ylabel('b array')
axes2.set_title('Axeas2 Title')
```
